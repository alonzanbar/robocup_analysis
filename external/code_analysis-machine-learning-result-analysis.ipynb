```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

model_file_name = '12many_1550349111'
```

```python
df_reduced,result_l = pickle.load(open('checkpoints/'+model_file_name + '.model'))
```

```python
pd.set_option('max_colwidth',100)
```

```python
#resluts_df = resluts_df.sort_values('f1_score',ascending=False)
res_df = result_l.loc[result_l['model_name']=='Boosted Trees',:].copy()
res_df['auc'] = 1- res_df['auc']
res_df = res_df.round(2)
```

```python
res_df = res_df.sort_values('f1_score',ascending=False).reset_index()
res_df['auc'] = 1 - res_df['auc']
```

```python
res_df['feature rank'] = res_df['feature rank'].apply(lambda x: [a[8:] for a in x][:3])
```

```python
res_df = result_l[result_l['model_name']=='Boosted Trees']
df_l = res_df.sort_values('auc',ascending=False)
lw=1
l = df_l.shape[0]
cols = 3
rows = 2
plt.figure(figsize=(12,8))
for i, res in enumerate(df_l.iterrows()):
    res  = res[1]
    plt.subplot(cols,rows,i+1)
    plt.plot(res['tpr'],res['fpr'],
         lw=lw, label="({0} {1} (area={2:.2f})".format(i,res['class 1'],res['auc'])) 
    plt.plot([0, 1], [0, 1], lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.title("({0}) {1} (area={2:.2f})".format(i,res['class 1'] + ' Vs ' + res['class 0'],res['auc']))
    #plt.legend(loc="upper left")
    plt.grid(True)
plt.tight_layout()
plt.savefig('fig/'+model_file_name+'_ml-auc-results.eps',dpi=1200,format='eps')
plt.show();
```

```python
from mlxtend.plotting import plot_decision_regions
from xgboost import XGBClassifier
from sklearn import linear_model
classification_series = result_l.iloc[6,:]
clf = classification_series['model']
seleted2= classification_series['feature rank'][:2]
```

```python
import copy
new_clf = copy.deepcopy(clf)
X = df_reduced[seleted2].values
Y = df_reduced['agent_type']==classification_series.type_class
Y = Y.values.astype(np.integer)
new_clf.fit(X,Y)
```

```python
plt.figure(figsize=(20,10))
with plt.style.context({'lines.markersize': 10}):
    ax = plot_decision_regions(X,Y,clf=new_clf)
L=plt.legend()
L.get_texts()[0].set_text('others')
L.get_texts()[1].set_text(classification_series.type_class)

# Adding axes annotations
plt.xlabel(seleted2[0])
plt.ylabel(seleted2[1])
plt.title(classification_series.model_name + ' Classification of agent types')
```

```python
import time

from sklearn.manifold import TSNE

n_sne = 7000

time_start = time.time()
tsne = TSNE(n_components=3, verbose=1, perplexity=100, n_iter=3000)
tsne_results = tsne.fit_transform(df_reduced.select_dtypes(include=[np.number]).values)

print 't-SNE done! Time elapsed: {} seconds'.format(time.time()-time_start)
```

```python
import seaborn as sns
df_reduced['x-tsne'] = tsne_results[:,0]
df_reduced['y-tsne'] = tsne_results[:,1]
fig,ax = plt.subplots(figsize=(16,10))
sns.scatterplot(ax=ax,data=df_reduced,x='x-tsne',y='y-tsne' ,hue='agent_type')
```
