# analyze using anomaly detection

```python
def line_to_map(line):
    return {i:line.values[i] for i in range (0, len(line.values))}
```

```python
def print_anomalies(anomaly_analysis):
    anomalies = anomaly_analysis.anomalies
    values = anomaly_analysis.time_series.values
    index = 0
    for anomaly in anomalies:
        start = anomaly.start_timestamp
        end = anomaly.end_timestamp
        exact = anomaly.exact_timestamp
        print(start, end, anomaly.anomaly_score, exact, values[start-1:end+2], "\n")
        index += 1
```

```python
def find_anomalies(line, is_only_decrease=False, is_to_print=False):
    anomaly_analysis = AnomalyDetector(line_to_map(line), score_threshold=1)
    if is_to_print:
        print_anomalies(anomaly_analysis)
    failures_markers_on = []
    filtered_failures_markers_on = []
    for anomaly in anomaly_analysis.anomalies:
        is_to_filter = False
        if is_only_decrease:
            if not(is_decrease_anomaly(anomaly, anomaly_analysis)):
                if is_to_print:
                    print("filter anomaly since not decrease {}".format(anomaly))
                is_to_filter = True
            elif not (is_continues_decrease_anomaly(anomaly, anomaly_analysis, is_to_print)):
                if is_to_print:
                    print("filter anomaly since not continues decrease {}".format(anomaly, anomaly.anomaly_score)) 
                is_to_filter = True
        if is_to_filter:
            filtered_failures_markers_on.append(anomaly.start_timestamp)
        else:
            failures_markers_on.append(anomaly.start_timestamp)
            
    return anomaly_analysis, failures_markers_on,filtered_failures_markers_on
```

```python
def plot_anomalies(line, is_only_decrease=False):
    anomaly_analysis, failures_markers_on,filtered_failures_markers_on = find_anomalies(line, is_only_decrease, True)
    print("Kept {} from {}".format(len(failures_markers_on), len(anomaly_analysis.anomalies)))
    print("The ones that kept - {}".format(failures_markers_on))
    y = line.values
    x = [i for i in range (0, len(y))]
    y_head = [y[i]+1 if i in filtered_failures_markers_on else 0 for i in range (0, len(y))]
    plt.plot(x, y, '-bD', markevery=failures_markers_on)
    plt.plot(x, y_head, '-gD', markevery=filtered_failures_markers_on)
    plt.show()
```

```python
def plot_anomalies1(line, is_only_decrease=False):
    anomaly_analysis = AnomalyDetector(line_to_map(line), score_threshold=1)
    print_anomalies(anomaly_analysis)
    failures_markers_on = []
    filtered_failures_markers_on = []
    for anomaly in anomaly_analysis.anomalies:
        is_to_filter = False
        if is_only_decrease:
            if not(is_decrease_anomaly(anomaly, anomaly_analysis)):
                print("filter anomaly since not decrease {}".format(anomaly))
                is_to_filter = True
            elif not (is_continues_decrease_anomaly(anomaly, anomaly_analysis, True)):
                print("filter anomaly since not continues decrease {}".format(anomaly, anomaly.anomaly_score)) 
                is_to_filter = True
        if is_to_filter:
            filtered_failures_markers_on.append(anomaly.start_timestamp)
        else:
            failures_markers_on.append(anomaly.start_timestamp)
    print("Kept {} from {}".format(len(failures_markers_on), len(anomaly_analysis.anomalies)))
    print("The ones that kept - {}".format(failures_markers_on))
    y = line.values
    x = [i for i in range (0, len(y))]
    y_head = [y[i]+1 if i in filtered_failures_markers_on else 0 for i in range (0, len(y))]
    plt.plot(x, y, '-bD', markevery=failures_markers_on)
    plt.plot(x, y_head, '-gD', markevery=filtered_failures_markers_on)
    plt.show()
```

```python
def is_decrease_anomaly(anomaly, anomaly_analysis):
    start = anomaly.start_timestamp
    end = anomaly.end_timestamp
    exact = anomaly.exact_timestamp
    values = anomaly_analysis.time_series.values
    return values[start-1] > values[start]
```

```python
def is_continues_decrease_anomaly(anomaly, anomaly_analysis, is_to_print):
    start = anomaly.start_timestamp
    end = anomaly.end_timestamp
    exact = anomaly.exact_timestamp
    values = anomaly_analysis.time_series.values
    #is_continues_decrease = values[start] >= values[start+1]
    if start == end:
        is_continues_decrease = values[start] >= values[start+1]
    else:
        if is_to_print:
            print(values[start+1:end])
        is_continues_decrease = values[start] >= values[end]
    #    is_continues_decrease = values[start] >= min(values[start:end])
    return is_continues_decrease
```
