```python
import collections
from collections import Counter
```

```python
def classify_one_vs_many(df,model_name,model,feature_to_class,type_class,type_0_class=None):   
    GH_df_reduced_one_vs_many = df.copy()
    
    category_df = GH_df_reduced_one_vs_many[GH_df_reduced_one_vs_many[feature_to_class]==type_class].copy()
    if type_0_class is None:
        others_df = GH_df_reduced_one_vs_many[((GH_df_reduced_one_vs_many[feature_to_class]!=type_class) & (
                                GH_df_reduced_one_vs_many['agent_type']=='non-agent'))].copy()
    else:
        others_df = GH_df_reduced_one_vs_many[(GH_df_reduced_one_vs_many['feature_to_class']==type_0_class) ].copy()
    
    y_false = [0] * others_df.shape[0]  
    y_true = [1] * category_df.shape[0]
    
    agent_type = category_df['agent_type'].unique()[0]
  
    df_merged = pd.concat([others_df,category_df],ignore_index=True)
    
    X = df_merged.select_dtypes(include=[np.number])
    y = np.array(y_false+y_true)
    test_size = 0.2
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size)
    if len(Counter(y_train)) <2:
        print ('cannot fit for {}'.format(type_class))
        return None
   
    if Counter(y)[type_class] == 1:
        print ('cannot fit for {}'.format(type_class))
        return
    
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    fpr = tpr = roc_auc =  None
    t=True
    try:
        y_pred = model.predict_proba(X_test)[:,0]
    except:
        t = False
    if t:
        fpr, tpr, _ = roc_curve(y_test, y_pred)
        roc_auc = auc(tpr, fpr)
    f1 = f1_score(y_test,pred)

    
    return {'model_name':model_name,'agent_typr':agent_type,'model':model, 'fpr':fpr, 'tpr':tpr, 'auc':roc_auc,'f1_score':f1,'class 0':type_0_class, 'class 1':type_class}
```

```python
df_reduced = pd.read_csv('checkpoints/ml-data.csv',index_col=0)
```

```python
from xgboost import XGBClassifier
from sklearn import linear_model
models = {'Boosted Trees':XGBClassifier}
```

```python
from sklearn import ensemble
categories = df_reduced['category'].unique()
results = []
for k,clf in models.iteritems():
    results += [classify_one_vs_many(df_reduced,k,clf(),'category',cat) for cat in categories]
    print "finished training : " +k 
    
```

```python
results = [res for res in results if res is not None]
df = pd.DataFrame(results)
X = np.array(df_reduced.columns)
```
