```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import re
```

```python
df_raw = pd.read_csv('data/merged-logs_repo.csv')
```

```python
df_code = df_raw[df_raw['is_source']==True]
```

```python
df_code.head().describe()
```

Number of repositories with source code that we could find relevant results file

```python
cor_df = df_code.corr()
```

```python
cor_df = df_code.corr()[[a for a in cor_df if 'results' in a]]
```

```python
cor_df = cor_df.filter(like='analizo', axis=0)
```

### pearson correlation between code metrics and results

The maximum values presented above ( absulote ) shows pearson correlation of all results metrics with the collected code metrics the restuls suggested no correlation 

```python
df_code_cols = df_code[[a for a in df_code.columns if re.search('analizo(.)*(mean|max|min)',a) ]].copy()

df_code_cols.loc[:,'results_possitive_dif'] = (df_code['results_win_diff'] >0)
```

```python
df_melted = df_code_cols.melt('results_possitive_dif',var_name='vars', value_name='val')

g = sns.catplot(x="results_possitive_dif",y = 'val',
                 col="vars",
                data=df_melted, kind="box",
                height=4, aspect=.7,col_wrap=5,sharey=False);
g.set_xticklabels(rotation=90)
```

```python
df_code_cols = df_code[[a for a in df_code.columns if re.search('analizo(.)*(mean|max|min)',a) ]].copy()

df_code_cols.loc[:,'results_goals_diff_bins'] = pd.qcut(df_code['results_goals_diff'],3,retbins=True,labels=['bad','mid','good'])[0]
```

```python
df_melted = df_code_cols.melt('results_goals_diff_bins',var_name='vars', value_name='val')

g = sns.catplot(x="results_goals_diff_bins",y = 'val',
                 col="vars",
                data=df_melted, kind="box",
                height=4, aspect=.7,col_wrap=5,sharey=False);
g.set_xticklabels(rotation=90)
```

```python
binary_df = df_raw[df_raw['is_source']!=False]
```

```python
binary_df = binary_df[[a for a in binary_df.columns if 'analizo' not in a and 'results' not in a]]
```

```python
import numpy as np
```

```python
binary_df= binary_df.select_dtypes(np.number)
```

```python
binary_df['is_rank'] = df_raw[df_raw['is_source']!=False]['results_rank'] >0
```

```python
binary_df_melt = binary_df.melt('is_rank',var_name='vars', value_name='val')

g = sns.catplot(x="is_rank",y = 'val',
                 col="vars",
                data=binary_df_melt, kind="box",
                height=4, aspect=1.5,col_wrap=2,sharey=False);
g.set_xticklabels(rotation=90)
```

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
fields = pd.read_csv('data/fields.csv')
```

```python
fields.columns = ['name','cat']
```

```python
binary_fields = fields[fields['cat'].isin(['basic-metric','radare','fast'])]
X = binary_df[[a for a in binary_fields['name'] if 'instructions' not in a]]
Y = binary_df['is_rank']
```

```python
X.loc[:,'year']  = df_raw['year']
X = X.drop('binary_coach_number_lines',axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.1,random_state=123)
model = xgb.XGBClassifier(scale_pos_weight= float(np.sum(~y_train)) / np.sum(y_train))
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
xgboost.plot_importance(model)
plt.rcParams['figure.figsize'] = [5, 5]
plt.show()
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
accuracy_score([False] * y_test.shape[0],y_test)
```
