```python
import matplotlib
import  pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import re
```

```python
df =  pd.read_csv("../data/data.csv",index_col=0)
```

```python
df.groupby('agent_type')['type'].count()
```

```python
categories = df['category'].unique()
```

```python
for cat in categories:
    fig= df[df['category']==cat][df['analizo_accm_mean']<100]['analizo_accm_mean'].hist()
    fig.set_title(cat)
    plt.show()
```

```python
df2 = df[~df['cccc_McCabes_cyclomatic_complexity'].isnull()].copy()
```

```python
df.groupby('agent_type')['cccc_lines_of_code','cccc_McCabes_cyclomatic_complexity'].corr()
```

```python

plot = sns.lmplot(height=10 , data = df2,x='cccc_lines_of_code',y='cccc_McCabes_cyclomatic_complexity',hue = 'agent_type')
ax = plt.gca()
ax.set_title('CC by loc small ranges (<100000)')
plt.show()
```

```python
df2['cc_to_loc_ratio'] = df['cccc_McCabes_cyclomatic_complexity'].divide(df['cccc_lines_of_code'])
```

```python
df2.groupby('category')['cc_to_loc_ratio'].mean().plot(kind='bar')
```
