```python
import matplotlib
import  pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import re
import pickle
```

```python
regexp_stat = re.compile(r'analizo_(.)*_(mean|sum)')
regexp_total = re.compile(r'analizo_total_(.)*')
rep_df  = pd.read_csv('data/3D-analysis-16_09_2018.csv')
code_df = rep_df[rep_df['is_source']==True]
model = pickle.load(open('dump'))
```

```python
code_df_reduced = code_df[[c for c in code_df.columns if regexp_stat.search(c) or regexp_total.search(c) or c == 'folder']].copy()
code_df_reduced['type'] = 'robo'
code_df_reduced['type_original'] = 'robo'
code_df_reduced.dropna(inplace=True)
```

```python
X = code_df_reduced.select_dtypes(include=[np.number])
model.predict(X)
```

```python
fg = sns.FacetGrid(data=code_df_reduced,  aspect=1.61,height=8)
fg.map(plt.scatter, 'analizo_accm_mean','analizo_acc_mean',).add_legend()
```
