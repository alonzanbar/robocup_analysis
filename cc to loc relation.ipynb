```python
import matplotlib
import  pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import re
```

```python
df =  pd.read_csv(os.getcwd() +"/data/data.csv",index_col=0)
df2 = df[~df['McCabes_cyclomatic_complexity'].isnull()].copy()
df2 = df2[df2['lines_of_code']>1000]
```

```python

plot = sns.lmplot(height=10 , data = df2[df2['lines_of_code']<10000],x='lines_of_code',y='McCabes_cyclomatic_complexity',hue = 'agent_type')
ax = plt.gca()
ax.set_title('CC by loc small ranges (<100000)')
plt.show()
```

```python
plot = sns.lmplot(height=10 , data = df2[df2['agent_type']=='agent'],x='lines_of_code',y='McCabes_cyclomatic_complexity',hue = 'agent_type')
ax = plt.gca()
ax.set_title('CC by loc agent type')
plt.show()
```

```python
plot = sns.lmplot(height=10 , data = df2[df2['agent_type']=='non-agent'],x='lines_of_code',y='McCabes_cyclomatic_complexity',hue = 'agent_type')
ax = plt.gca()
ax.set_title('CC by loc non agent type')
plt.show()
```
