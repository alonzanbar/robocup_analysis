```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =  pd.read_csv('data/robil_19_09_2018code_analysis_detailed_report.csv')
```

```python
df['module'] = df['analizo__filename'].str.split('/',2).str[1]
```

```python
#df.groupby('module').agg(['count','mean'])['analizo_accm'].plot(kind='barh',figsize=(10,15))
group_df = df.groupby('module').agg({'analizo_loc':['sum'],'analizo_accm':['mean']})[['analizo_accm','analizo_loc']]
```

```python
group_df.columns = group_df.columns.get_level_values(0)
```

```python
# f, ax = plt.subplots(figsize=(15, 15))
# g = sns.barplot(data=group_df,y='count',x = group_df.index)
# ax2 = plt.twinx()
# sns.barplot(data=group_df,y='mean',x = group_df.index, ax=ax2)
# for item in g.get_xticklabels():
#     item.set_rotation(90)

g = group_df.plot( kind= 'bar' , secondary_y= 'analizo_accm' , rot= 0,figsize=(15,8) )
for item in g.get_xticklabels():
     item.set_rotation(90)
```
